```python
import cv2
import numpy as np
from tkinter import *
from PIL import Image, ImageTk

# Load an image
image_path = 'girlBodBgRemoved.png'  # Replace with your image path
original_image = cv2.imread(image_path)
image = original_image.copy()

# Function to resize the image
def resize_body_parts(scale_leg, scale_hand, scale_neck, scale_shoulder, scale_width):
    global image
    image = original_image.copy()
    height, width = image.shape[:2]

    # Define the vertical regions for each body part
    shoulder_height = int(height * 0.2)
    neck_height = int(height * 0.1)
    hand_height = int(height * 0.3)
    leg_height = int(height * 0.25)
    bust_to_waist_start = shoulder_height + neck_height
    bust_to_waist_end = bust_to_waist_start + int(height * 0.3)

    # Extract and resize each part
    shoulder_part = image[:shoulder_height, :]
    shoulder_part_resized = cv2.resize(shoulder_part, (int(width * scale_width), int(shoulder_height * scale_shoulder)))

    neck_part = image[shoulder_height:shoulder_height + neck_height, :]
    neck_part_resized = cv2.resize(neck_part, (int(width * scale_width), int(neck_height * scale_neck)))

    hand_part = image[shoulder_height + neck_height:bust_to_waist_start, :]
    hand_part_resized = cv2.resize(hand_part, (int(width * scale_width), int(hand_height * scale_hand)))

    bust_to_waist_part = image[bust_to_waist_start:bust_to_waist_end, :]
    bust_to_waist_part_resized = cv2.resize(bust_to_waist_part, (int(width * scale_width), bust_to_waist_part.shape[0]))

    leg_part = image[height - leg_height:, :]
    leg_part_resized = cv2.resize(leg_part, (int(width * scale_width), int(leg_height * scale_leg)))

    # Calculate the new heights for each part
    new_shoulder_height = shoulder_part_resized.shape[0]
    new_neck_height = neck_part_resized.shape[0]
    new_hand_height = hand_part_resized.shape[0]
    new_bust_to_waist_height = bust_to_waist_part_resized.shape[0]
    new_leg_height = leg_part_resized.shape[0]

    # Calculate the total new height for the new image
    new_image_height = new_shoulder_height + new_neck_height + new_hand_height + new_bust_to_waist_height + new_leg_height
    new_image_width = max(shoulder_part_resized.shape[1], neck_part_resized.shape[1], hand_part_resized.shape[1], bust_to_waist_part_resized.shape[1], leg_part_resized.shape[1])

    # Create the new blank image
    new_image = np.zeros((new_image_height, new_image_width, 3), dtype=np.uint8)

    # Place the resized parts in the new image
    current_y = 0
    new_image[current_y:current_y + new_shoulder_height, :shoulder_part_resized.shape[1]] = shoulder_part_resized
    current_y += new_shoulder_height

    new_image[current_y:current_y + new_neck_height, :neck_part_resized.shape[1]] = neck_part_resized
    current_y += new_neck_height

    new_image[current_y:current_y + new_hand_height, :hand_part_resized.shape[1]] = hand_part_resized
    current_y += new_hand_height

    new_image[current_y:current_y + new_bust_to_waist_height, :bust_to_waist_part_resized.shape[1]] = bust_to_waist_part_resized
    current_y += new_bust_to_waist_height

    new_image[current_y:current_y + new_leg_height, :leg_part_resized.shape[1]] = leg_part_resized

    image = new_image
    update_image_display()

# Function to update the image display
def update_image_display():
    global image, photo
    b, g, r = cv2.split(image)
    img = cv2.merge((r, g, b))
    img = Image.fromarray(img)
    img = ImageTk.PhotoImage(image=img)
    canvas.create_image(0, 0, anchor=NW, image=img)
    canvas.image = img

# Create the main window
root = Tk()
root.title("Body Part Resizer")

# Create canvas to display image
canvas = Canvas(root, width=original_image.shape[1], height=original_image.shape[0])
canvas.pack()

# Create sliders for each body part
scale_leg = Scale(root, from_=0.5, to=1.5, resolution=0.1, orient=HORIZONTAL, label="Legs", command=lambda val: resize_body_parts(scale_leg.get(), scale_hand.get(), scale_neck.get(), scale_shoulder.get(), scale_width.get()))
scale_leg.set(1)
scale_leg.pack()

scale_hand = Scale(root, from_=0.5, to=1.5, resolution=0.1, orient=HORIZONTAL, label="Torso", command=lambda val: resize_body_parts(scale_leg.get(), scale_hand.get(), scale_neck.get(), scale_shoulder.get(), scale_width.get()))
scale_hand.set(1)
scale_hand.pack()

scale_neck = Scale(root, from_=0.5, to=1.5, resolution=0.1, orient=HORIZONTAL, label="Upper Torso", command=lambda val: resize_body_parts(scale_leg.get(), scale_hand.get(), scale_neck.get(), scale_shoulder.get(), scale_width.get()))
scale_neck.set(1)
scale_neck.pack()

scale_shoulder = Scale(root, from_=0.5, to=1.5, resolution=0.1, orient=HORIZONTAL, label="Head", command=lambda val: resize_body_parts(scale_leg.get(), scale_hand.get(), scale_neck.get(), scale_shoulder.get(), scale_width.get()))
scale_shoulder.set(1)
scale_shoulder.pack()

scale_width = Scale(root, from_=0.5, to=1.5, resolution=0.1, orient=HORIZONTAL, label="Width", command=lambda val: resize_body_parts(scale_leg.get(), scale_hand.get(), scale_neck.get(), scale_shoulder.get(), scale_width.get()))
scale_width.set(1)
scale_width.pack()

# Initialize image display
update_image_display()

# Start the GUI loop
root.mainloop()
```
